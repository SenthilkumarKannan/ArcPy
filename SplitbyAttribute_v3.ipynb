```python
import arcpy
import os

# Set workspace (folder where the input and output geodatabases are located)
arcpy.env.workspace = r"C:\Users\skannan\Desktop\temp"

# Set input feature class
input_fc = r"C:\Users\skannan\Desktop\temp\MasterAttribute_upto2024.gdb\Attribute_points"

# Set output geodatabase path
output_gdb = r"C:\Users\skannan\Desktop\temp\Attribute_Split.gdb"

# Set field to split by
split_field = "FeatureDescription"

# Check if output geodatabase exists, if not, create it
if not arcpy.Exists(output_gdb):
    arcpy.CreateFileGDB_management(os.path.dirname(output_gdb), os.path.basename(output_gdb))
    print(f"Created output geodatabase at {output_gdb}")

# Use Split By Attributes tool to split the input feature class
arcpy.SplitByAttributes_analysis(input_fc, output_gdb, split_field)

# List all feature classes created in the output geodatabase
arcpy.env.workspace = output_gdb
feature_classes = arcpy.ListFeatureClasses()

# Loop through each feature class and export it to the same output geodatabase
for fc in feature_classes:
    # Get the base name of the feature class
    base_name = os.path.splitext(fc)[0]
    
    # Check if the feature class already exists in the output geodatabase
    if not arcpy.Exists(os.path.join(output_gdb, base_name)):
        # Export the feature class to the same output geodatabase
        arcpy.FeatureClassToFeatureClass_conversion(fc, output_gdb, base_name)
        print(f"Exported {fc} to {output_gdb}")
    else:
        print(f"Feature class {base_name} already exists in {output_gdb}. Skipping.")

print("Feature classes exported successfully.")

```
