```python
import arcpy

# Set workspace (replace with your actual workspace path)
arcpy.env.workspace = r"C:\Senthil\14-3D Model_GC27\GC-27_RemovedEmptyFC.gdb"

# Get List of Datasets in Workspace
dataset_list = arcpy.ListDatasets("*", "Feature")

# Print and log messages
print("\n")
print("--------------------------------------------------")

# Loop through each dataset
for dataset in dataset_list:
    try:
        print(f"Processing dataset: {dataset}")
        
        # Skip specific dataset types
        desc = arcpy.Describe(dataset)
        if desc.datasetType in ["NetworkDataset", "UtilityNetwork", "TraceNetwork"]:
            print(f"Skipping dataset {dataset} (Type: {desc.datasetType})")
            continue
        
        # Get list of feature classes in the dataset
        feature_classes = arcpy.ListFeatureClasses("*", "", dataset)
        feature_classes.sort()

        # Loop through each feature class
        for fc in feature_classes:
            print(f"Processing feature class: {fc}")
            
            # Get record count
            record_count = int(arcpy.management.GetCount(fc)[0])
            
            if record_count == 0:
                print(f"Empty - Deleting {fc}")
                arcpy.management.Delete(fc)
            else:
                print(f"{record_count} Records - Keeping {fc}")

    except Exception as e:
        print(f"An error occurred while processing dataset {dataset}: {e}")

print("--------------------------------------------------")
print("\n")
```
